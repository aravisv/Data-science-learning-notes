```python
"""
for a simple linear regression, the aim is to fund those coefficients for which the loss function is minimum
different methods available are OLS, normal equation, gradient descent...
"""
```

```python
# generate data
import numpy as np

np.random.seed(0)

x = np.linspace(0,20,1000)
noise = np.random.normal(0,2,1000)

y = 3*x + 5 + noise

# regression formula will be of the form - y' = b0' + b1'x
```

```python
"""
OLS formula is 
b1' = (Sum (xi - mean_x ) * (yi - mean_y)) / Sum (Sqr (xi - mean_x))
b0' = mean_y - ( b1' * mean_x )
"""
mean_x = np.mean(x)
mean_y = np.mean(y)
b1 = np.sum ((x - mean_x) * (y - mean_y)) / np.sum( (x - mean_x)**2 )
b0 = mean_y - (b1 * mean_x)
[b0,b1]
```

```python
# same calculations using statsmodel
import statsmodels.api as sm
X = sm.add_constant(x) # OLS requires a constant term
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
"""
gradient descent
start with zero value for weights and bias
calculate the weights' and bias gradient 
adjust the weight to the opposite direction of the slope, by alpha
perform the steps until n number of iterations

X - feature matrix - (m, n)
y - target vector - (m, )
weights (w) - model parameters - (n, )
bias (b) - intercept - scalar (1)

"""
```

```python
def gradient_descent(X, y, num_iter, alpha):
    m, n = X.shape
    weights = np.zeros(n)
    bias = 0
    
    for i in range(num_iter):
        y_pred = X @ weights + bias    
        error = y_pred - y
    
        weight_gradient = 2 / m * np.dot(X.T, error)
        bias_gradient = 2 / m * np.sum(error)
    
        weights = weights - (alpha * weight_gradient)
        bias = bias - (alpha * bias_gradient)

    return [weights, bias]
```

```python
# shape the X and y properly to input into the gradient_descent function
X = x.reshape(-1,1)
y = y.flatten()

gradient_descent(X, y, 100, 0.001)
```
